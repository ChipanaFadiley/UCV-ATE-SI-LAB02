**1. Instalación e importación**

```python
# instalaciòn de opencv
!pip install opencv-python
# importaciòn de librerìas
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

**2. Carga y visualización de una imagen**


```python
# cargar imagen
uploaded = files.upload()
# leer y mostrar imagen
imagen = cv2.imread('imagen_test.png')
imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

plt.imshow(imagen_rgb)
plt.title("IMAGEN ORIGINAL")
plt.axis("off")
plt.show()
```

**3. Conversión a escala de grises**

```python
# conversiòn a escala de grises
gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

plt.imshow(gris, cmap='gray')
plt.title("IMAGEN EN ESCALA DE GRISES")
plt.axis("off")
plt.show()
```

**4. Modificaciòn de Valores**

```python
brillo = cv2.convertScaleAbs(gris, alpha=1.3, beta=30)

plt.imshow(brillo, cmap='gray')
plt.title("IMAGEN CON BRILLO AJUSTADO")
plt.axis("off")
plt.show()
```

**5. Cambio de valores en alpha y beta**

```python
brillo = cv2.convertScaleAbs(gris, alpha=3, beta=2)

plt.imshow(brillo, cmap='gray')
plt.title("IMAGEN CON BRILLO AL MÀXIMO")
plt.axis("off")
plt.show()
```

**6. Detecciòn de bordes**

```python
# detector de bordes Canny
bordes = cv2.Canny(imagen, 100, 200)

plt.imshow(bordes, cmap='gray')
plt.title("DETECCIÒN DE BORDES")
plt.axis("off")
plt.show()
```
